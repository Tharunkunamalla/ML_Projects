```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("canada_per_capita_income.csv")
```

```python
print(data.head())
```

```python
print(data.tail())
```

```python
X = data[['year']]
Y = data[['per capita income (US$)']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print(f"Slope(m) : ", model.coef_[0])
print(f"Intercept(c) : ", model.intercept_)
```

```python
y_predic = model.predict(x_test).flatten()   #actual and predicted are arrays with one element, so actual[0] extracts the scalar value.
y_actual = y_test.values.flatten()

for actual, predicted in zip(y_actual, y_predic):
    print(f"Actual: {actual:.2f} , Predicted: {predicted:.2f}")  #This way, the formatting :.2f works on scalars.
```

```python
# Plotting the regression line over the data
plt.scatter(X,Y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.xlabel('Year')
plt.ylabel('Percapita-Income')
plt.title('Year vs Percapita-Income')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print("Mean Squared Error:", mean_squared_error(y_test, y_predic))
print("R-squared Score:", r2_score(y_test, y_predic))
```

```python
Years = pd.DataFrame({'year': [2030]})  # Same column name as training
predicted_cost = model.predict(Years)

print("Predicted cost:", predicted_cost[0])
```
