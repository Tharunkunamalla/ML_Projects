```python
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score #cross validation data and trainin data gets divided in iterations
%matplotlib inline
```

```python
df = fetch_california_housing()
```

```python
df
```

```python
pd.DataFrame(df.data)
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names
print(dataset.head())
```

```python
dataset
```

```python
dataset.head()
```

```python
dataset['price'] = df.target
```

```python
dataset
```

```python
#Dividing the dataset into independent and dependent features

X = dataset.iloc[:,:-1]
Y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split


x_train, x_test,y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state = 42)
```

```python
#Linear Regression
model = LinearRegression()
model.fit(x_train, y_train)
mse = cross_val_score(model, X, Y, scoring = 'neg_mean_squared_error', cv = 5)
print(mse)          #CV -- > cross validation done is 5 times in a loop
```

```python
mean_mse = np.mean(mse)
print(mean_mse)  #negative mean squared
```

```python
#model.predict()
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()

param = {'alpha' : [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20]}  # takes a alpha value which suits 
ridge_regressor = GridSearchCV(ridge, param,scoring = 'neg_mean_squared_error', cv =5)

ridge_regressor.fit(X,Y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
#if best score in ridge is -32 and  mean_mse in linear regression -37 then use linear bcz performing well
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
# Define model and parameters
lasso = Lasso()  # Increase max_iter to avoid convergence warnings
param = {'alpha': [1e-4, 1e-3, 1e-2, 1, 10, 20]}

# Grid search with 5-fold cross-validation
lasso_regressor = GridSearchCV(lasso, param, scoring='neg_mean_squared_error', cv=5)

# Fit model
lasso_regressor.fit(X, Y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```
